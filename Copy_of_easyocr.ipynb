```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

# Read in images

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
IMAGE_PATH=("/content/drive/MyDrive/surf.jpeg")
```

```python
reader= easyocr.Reader(['en'],gpu=False)
result=reader.readtext(IMAGE_PATH)
result
```

# Draw Results

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

# Handling Multiple Lines PART

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
PATH=("/content/drive/MyDrive/sign.png")
```

```python
reader= easyocr.Reader(['en'],gpu=False)
result=reader.readtext(PATH)
result
```

```python
res = reader.readtext('/content/drive/MyDrive/sign.png',detail=0,paragraph=True)
print(res)
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(PATH)
spacer = 150
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(255,0,0),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
